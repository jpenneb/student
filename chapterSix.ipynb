```python
# histogram but better

import graphics as gr

def appendings(number, skre, num):
    if skre == num:
        number.append(skre)

def main():
    
    win = gr.GraphWin('Histogram')
    win.setCoords(0,0,100,5)
    file = open('histogram.txt', 'r')

    one = []
    two = []
    three = []
    four = []
    five = []
    six = []
    seven = []
    eight = []
    nine = []
    ten = []
    
    for score in file.readlines():
        score = int(score)
        appendings(one, score, 1)
        appendings(two, score, 2)
        appendings(three, score, 3)
        appendings(four, score, 4)
        appendings(five, score, 5)
        appendings(six, score, 6)
        appendings(seven, score, 7)
        appendings(eight, score, 8)
        appendings(nine, score, 9)
        appendings(ten, score, 10)
        
    forPlotting = [sum(one)/1,sum(two)/2,sum(three)/3,sum(four)/4,sum(five)/5,sum(six)/6,
          sum(seven)/7,sum(eight)/8,sum(nine)/9,sum(ten)/10]
    
    x = 10
    for plot in forPlotting:
        bar = gr.Rectangle(gr.Point(x,0),gr.Point(x+4,plot))
        bar.draw(win)
        x = x + 8
    
main()
```

```python
#old mcdonald

def oldMacDonald():
    print("Old MacDonald had a farm, Ee-igh, Ee-igh Oh!")

def animalLyrics(animal, sound):
    print("And on that farm he had a", animal, "Ee-igh, Ee-igh, Oh!")
    print("With a", sound + ",", sound, "here and a", sound + ",", sound, "there.")
    print("Here a", sound, "there a", sound, ", everywhere a", sound + ",", sound, ".")

def fullMacDonaldSong():
    oldMacDonald()
    animalLyrics("cow", "moo")
    oldMacDonald()
    print()
    oldMacDonald()
    animalLyrics("chicken", "cluck")
    oldMacDonald()

fullMacDonaldSong()    
```

```python
#sum N and sum N cubes

def sumN(n):
    lists = []
    for i in range(n+1):
        lists.append(i)
    return sum(lists)
    
def sumNCubes(n):
    lists = []
    for i in range(n+1):
        lists.append(i**3)
    return sum(lists)

def main():
    num = eval(input("Give me a number! "))
    summ = sumN(num)
    summCube = sumNCubes(num)
    
    print(summ,",", summCube)
    
main()
```

```python
# square list from file separated by commas

def stringListToNumList(stringList):
    newNums = []
    for num in stringList:
        newNum = eval(num)
        newNums.append(newNum)
    return newNums    

def squareEach(numList):
    i = 0
    for num in numList:
        numList[i] = numList[i]**2
        i = i + 1

def sumList(summerList):
    return sum(summerList)
        
def main():
    fileName = input("Provide the file of a list with numbers: ")
    file = open(fileName, 'r')
    listt = file.readline().split(',')
    numList = stringListToNumList(listt)
    squareEach(numList)
    summedList = sumList(numList)
    print(summedList)
main()
```

```python
import graphics as gr

def drawHead(win, center, size):
       
    headPoint1 = center.clone()
    headPoint1.move(-80.75,-110.5)
    
    headPoint2 = center.clone()
    headPoint2.move(80.75,110.5)
    
    head = gr.Oval(gr.Point(headPoint1.getX()-size,headPoint1.getY()-size),
                   gr.Point(headPoint2.getX()+size,headPoint2.getY()+size))
    head.setFill('tan')
    head.draw(win)
    
def drawEyes(win, center):
    
    leftEyeLoc = center.clone()
    leftEyeLoc.move(-25,25)
    
    leftEye = gr.Circle(leftEyeLoc, 15)
    leftEye.draw(win)
    leftEye.setFill('white')
    pupilL = leftEyeLoc
    pupilL.draw(win)
    eyebrowL = gr.Line(gr.Point(leftEyeLoc.getX()-20,leftEyeLoc.getY()+20),
                       gr.Point(leftEyeLoc.getX()+15,leftEyeLoc.getY()+20))
    eyebrowL.draw(win)

    rightEyeLoc = center.clone()
    rightEyeLoc.move(25,25)
    
    rightEye = gr.Circle(rightEyeLoc, 15)
    rightEye.draw(win)
    rightEye.setFill('white')
    pupilR = rightEyeLoc
    pupilR.draw(win)
    eyebrowR = gr.Line(gr.Point(rightEyeLoc.getX()-15,rightEyeLoc.getY()+30),
                       gr.Point(rightEyeLoc.getX()+20,rightEyeLoc.getY()+20))
    eyebrowR.draw(win)  
    
def drawNose(win, center):
    line1 = gr.Line(gr.Point(center.getX(),center.getY()+10),
                    gr.Point(center.getX()+10,center.getY()-10))
    line1.draw(win)
    line2 = gr.Line(gr.Point(center.getX()+10,center.getY()-10),
                    gr.Point(center.getX(),center.getY()-10))
    line2.draw(win)

def drawMouth(win, center):
    mouth = gr.Circle(gr.Point(center.getX()-15,center.getY()-25) , 10)
    mouth.draw(win)
    mouth.setFill('black')
    mouth.setOutline('maroon')    
    
def drawFace(win, center, size):
   
    drawHead(win, center, size)
    drawEyes(win, center)
    drawNose(win, center)
    drawMouth(win, center)

def main():
    
    photo = gr.Image(gr.Point(0,0), 'C:/Users/Jarrett/Downloads/giphy.gif')
    win = gr.GraphWin('image', photo.getWidth(), photo.getHeight())
    
    wid, hei, anc = photo.getWidth(), photo.getHeight(), photo.getAnchor()
    
    win.setCoords(anc.getX()-wid,anc.getY()-hei,wid,hei)
    photo.draw(win)
    
    faces = eval(input("How many faces? "))
    
    for i in range(faces):
        size = eval(input("How big (>0)? "))
        center = win.getMouse()        
        drawFace(win, center, size)
main()
```

```python
import graphics as gr

def determineShape(initialP, shape):
    if shape == 'Circle':
        circ = gr.Circle(gr.Point(initialP.getX(),initialP.getY()), 20)
        return circ
    
    if shape == 'Square':
        sqr = gr.Rectangle(gr.Point(initialP.getX()-10,initialP.getY()-10), gr.Point(initialP.getX()+10,initialP.getY()+10))
        return sqr
    
    if shape == 'Triangle':
        trian = gr.Polygon(gr.Point(initialP.getX()-10,initialP.getY()-10), gr.Point(initialP.getX()+10,initialP.getY()-10), gr.Point(initialP.getX()+10,initialP.getY()+10))
        return trian

def moveTo(newCenter, zeShape, shape):

    if shape == 'Circle':
        pointOfCenter = zeShape.getCenter()
        zeShape.move(newCenter.getX()-pointOfCenter.getX(), newCenter.getY()-pointOfCenter.getY())
        return zeShape

    elif shape == 'Square':
        aPoint = zeShape.getP1()
        zeShape.move(newCenter.getX()-aPoint.getX(), newCenter.getY()-aPoint.getY())
        return zeShape
    
    if shape == 'Triangle':
        aPoint = zeShape.getPoints()[0]
        zeShape.move(newCenter.getX()-aPoint.getX(), newCenter.getY()-aPoint.getY())
        return zeShape
    
def main():
    
    win = gr.GraphWin('shapes', 150, 150)
    shape = input("What shape would you like to draw? Circle, Square, or Triangle ")
    
    theShape = 0

    for i in range(10):
        if theShape == 0:
            initialPoint = win.getMouse()
            theShape = determineShape(initialPoint, shape)
            theShape.draw(win)
        else:
            newPoint = win.getMouse()
            theShape = moveTo(newPoint, theShape, shape)
    
main()
```
